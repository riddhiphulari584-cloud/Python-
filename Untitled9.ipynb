```python
# WAP to demonstrate type casting in Python

a = 10        # integer
b = 3.5       # float

# Type casting
c = float(a)  # converting int to float
d = int(b)    # converting float to int
e = str(a)    # converting int to string

print("Integer value:", a, "Type:", type(a))
print("Float value:", b, "Type:", type(b))
print("After casting int to float:", c, "Type:", type(c))
print("After casting float to int:", d, "Type:", type(d))
print("After casting int to string:", e, "Type:", type(e))
```
