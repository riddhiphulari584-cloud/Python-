```python
# Creating a tuple
t = (10, 20, 30, 40, 50)

# Indexing
print("Element at index 0:", t[0])

# Negative Indexing
print("Element at last index:", t[-1])

# Slicing
print("Elements from index 1 to 3:", t[1:4])

# Iteration
print("Iterating elements:")
for i in t:
    print(i)
```
