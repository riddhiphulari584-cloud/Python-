```python
# WAP to demonstrate logical operators

a = 10
b = 5

print("a =", a)
print("b =", b)

# Logical AND
print("a > 5 and b > 2:", a > 5 and b > 2)

# Logical OR
print("a > 15 or b > 2:", a > 15 or b > 2)

# Logical NOT
print("not(a > 5):", not(a > 5))
```
